```python
import sys
sys.path.append("../src")

from classifier import train_classifier

# Train on larger dataset
model, label_map = train_classifier("../data/franglais_samples_large.csv")

from embeddings import embed_texts
from preprocessing import preprocess_text

def predict(text):
    clean = preprocess_text(text)  # no lang param
    emb = embed_texts([clean])
    pred = model.predict(emb)[0]

    inv_map = {v:k for k,v in label_map.items()}
    return inv_map[pred]

# Test predictions
test_sentences = [
    "I love ce restaurant",
    "Aujourd'hui I go beach",
    "Je mange du riz",
    "I am eating food",
    "Ce laptop is broken",
    "Je go to the gym"
]

for s in test_sentences:
    print(s, "→", predict(s))
```
