```python
import sys
import numpy as np
sys.path.append("../src")

from preprocessing import preprocess_text
from embeddings import embed_texts

# Example English & French sentences
sentences = [
    "Machine learning is transforming computer science.",
    "L'apprentissage automatique transforme l'informatique."
]

# Preprocess
cleaned = [
    preprocess_text(sentences[0], lang="en"),
    preprocess_text(sentences[1], lang="fr")
]

print("Cleaned:", cleaned)

# Get embeddings
emb = embed_texts(cleaned)

print("Embedding shape:", emb.shape)


sim = np.dot(emb[0], emb[1]) / (np.linalg.norm(emb[0]) * np.linalg.norm(emb[1]))

print("Similarity:", sim)
```
